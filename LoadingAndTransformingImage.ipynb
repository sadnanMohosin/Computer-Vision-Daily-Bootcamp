```python
import skimage
import cv2
import numpy as np

from PIL import Image, ImageFilter 
from sklearn.feature_extraction import image 

import matplotlib.pyplot as plt 
```

```python
img_stork = Image.open('bird.jpg')
```

```python
plt.figure(figsize=(6,12))

plt.imshow(img_stork)
plt.title('Original image')
plt.show()
```

```python
img_stork.format,img_stork.mode,img_stork.size     
```

```python
#gray-scale image 
gs_image = img_stork.convert(mode='L')
# gs_image = img_stork.convert(mode='P')
gs_image
```

```python
gs_image.size #image size reamin unchanged  
```

```python
#lager image size hard to work with
gs_image.thumbnail((200,200))

gs_image
```

```python
gs_image.size , img_stork.size
```

```python
#another funtion to resie image 
#its will not maintain aspect ration
img_resize = img_stork.resize((300,300))
img_resize
```

```python
hoz_flip = img_stork.transpose(Image.FLIP_LEFT_RIGHT)
ver_flip = img_stork.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
plt.figure(figsize=(12,6))

plt.title ("Horizontal Flip")
plt.imshow(hoz_flip)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.title ("Vertical Flip")
plt.imshow(ver_flip)

plt.show()
```

```python
#python library alow us to work with array image 
img_stork_arr = np.array(img_stork)

type(img_stork_arr)
```

```python
img_stork_arr.shape
```

```python
#we can visualize our numpy array image 
plt.figure(figsize=(6,6))
plt.title("Numpy Image")
plt.imshow(img_stork_arr,cmap='gray')
```

```python
from skimage.color import rgb2gray

img_stork_arr_gray = rgb2gray(img_stork_arr)

plt.figure(figsize=(6,6))
plt.title("Gray Image")
plt.imshow(img_stork_arr_gray,cmap='gray')
```

```python
from skimage.transform import resize
stork_resized = resize(img_stork_arr,(224,224),anti_aliasing=True)
# setting anti aliasing to "True" will reduce the distortions that might be introduced in your resized lower resoution image
stork_resized.shape
```

```python
plt.figure(figsize=(6,6))
plt.title("Resized Image")
plt.imshow(stork_resized,cmap='gray')
```
